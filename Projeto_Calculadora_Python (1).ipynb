# Calculadora Python

```python
# Recebendo os dados.
nome = input('Digite seu nome: ')
print('Bem vindo ', nome, 'a calculadora!')
num1 = input('Digite o primeiro número: ')
num2 = input('Digite o segundo número: ')
opr = input('Qual operação deseja realizar (+,-,*,/): ')

# Convertendo os dados recebidos
num1 = int(num1)
num2 = int(num2)

# Operações Lógicas

if opr == '+':
  print('A soma dos dois numeros é: ', num1 + num2)
elif opr == '-':
    print('A subtração dos dois números é: ', num1 - num2)
elif opr == '*':
  print('A multiplicação dos dois números é: ', num1 * num2)
elif opr == '/':
  print('A divisão dos dois números é: ', num1 / num2)
```
